# Benchmark Lalande Transformer (Lalande et al.)

**Setup:**
- 1. conda create --name tf4sr_env --file tf4sr_env.txt
- 2. conda activate tf4sr_env

```python
from lalande_regressor import TransformerLalande
import numpy as np

X = np.random.rand(100, 2)
y = X[:, 0]+X[:, 1]

regr = TransformerLalande()
regr.fit(X, y)
print(regr.model())
pred = regr.predict(X)
np.mean((y - pred)**2)
```

```python
# Test Regressor Interface
import numpy as np

from nesymres_regressor import TransformerBiggio

X = np.random.rand(100, 2)
y = X[:, 0]+X[:, 1]

print("X shape: ", X.shape)
print("y shape: ", y.shape)

regr = TransformerBiggio()
regr.fit(X, y)
print(regr.model())
pred = regr.predict(X)
np.mean((y - pred)**2)
```
